## leetcode206 反转链表
给你单链表的头节点 head ，请你反转链表，并返回反转后的链表

<font color = 'blue'>反转链表可以通过**迭代**和**递归**两种方式进行</font>

**迭代**：
迭代的主要思想是使用前一个节点prev，以及当前节点curr进行递归处理

```python
def AreverseList(head):
        if not head or not head.next:
            return head
        
        prev = None
        curr = head
        while curr:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        return prev
```

**递归**：
本题中，newhead在找到最后一个节点便固定下来。在**回溯**的时候进行每个节点的反转操作。

```python
def reverseList(head):
        if not head or not head.next:
            return head
        
        new_head = reverseList(head.next)

        head.next.next = head
        head.next = None

        return new_head
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def printList(head):
    curr = head
    while curr:
        if curr.next:
            print(curr.val,end = '->')
        else:
            print(curr.val,end = '')
        curr=curr.next
    print('')
head1 = ListNode(1,ListNode(2,ListNode(3,ListNode(4,ListNode(5)))))
head2 = ListNode(1,ListNode(2,ListNode(3,ListNode(4,ListNode(5)))))
head3 = ListNode(1,ListNode(2,ListNode(3,ListNode(4,ListNode(5)))))
printList(head1)
printList(AreverseList(head2))
printList(reverseList(head3))
```

## leetcode92 反转链表2
给你单链表的头指针 head 和两个整数 left 和 right ，其中 left <= right 。请你反转从位置 left 到位置 right 的链表节点，返回 反转后的链表 。
<font color = 'blue'>本题主要有两种解法，第一种是找到反转的区域，将其切断，然后使用leetcode206的方法反转，反转后将其连接回链表。也可以采用**穿针引线法/头插法**一次遍历解决，以下展示穿针引线法的使用</font>

```python
def reverseBetween(head, left, right):
    if not head.next:
        return head
    
    count = 1
    DummyHead = ListNode(-1,head)
    prev = DummyHead
    curr = head
    for i in range(left-1):
        prev = prev.next
        

    curr = prev.next

    for i in range(right - left):
        next_node = curr.next
        curr.next = next_node.next
        next_node.next = prev.next #此处一定是prev.next而不是curr,因为curr在操作中在不停向前移动.详见下图流程
        prev.next = next_node
    return DummyHead.next
```

其主要思想为不停将**curr节点后的节点**插入到区间头部，因此也叫做**头插法**

## Leetcode86 分隔链表
给你一个链表的头节点 head 和一个特定值 x ，请你对链表进行分隔，使得所有 小于 x 的节点都出现在 大于或等于 x 的节点之前。
你应当 保留 两个分区中每个节点的初始相对位置。
示例 1：
输入：head = [1,4,3,2,5,2], x = 3
输出：[1,2,2,4,3,5]

示例 2：
输入：head = [2,1], x = 2
输出：[1,2]

官方题解为维护两个新链表储存小于和大于的数，然后将其连接。此外，本题还可以使用**尾插法**的方式来解决。

```python
def partition(head, x):
    if not head or not head.next:
        return head
    DummyHead = ListNode(-1,head)
    curr = head
    n = 1
    while curr.next:
        n+=1
        curr = curr.next
    tail = curr

    curr = head
    prev = DummyHead
    while n>0:
        n-=1
        if curr.val>=x:
            if curr.next:
                next_node = curr.next
                prev.next = next_node
                tail.next = curr
                tail = curr
                curr.next = None
                curr = next_node
        else:
            prev = curr
            curr = curr.next

    return DummyHead.next
```
