# 对于各类数据结构使用的总结

## 哈希表
在python中，可以使用list和dict实现哈希表。使用dict更为方便。<font color = 'red'>可以直接用if char in dict进行查找</font>hot100-1  
dict的键必须是不可变对象。不能使用list当作dict的键，但是可以使用tuple。使用tuple作为dict，即哈希表的键，可以实现字母异位词的分组。hot100-2  
字典的dict.keys() dict.values()返回的都是字典视图对象，使用时最好转换为list  
使用list实现哈希表有一个主要问题是key值不能为负。key有负值时使用set和dict。其中set可以很好解决重复值的情况。

## 双指针
可以用for循环控制右指针，避免使用while进行循环操作，逻辑更清晰。  
list.sort()会对列表原地修改，无返回值。如果需要，可以使用sort_list = sorted(list)。  

## 滑动窗口
滑动窗口可以算是双指针的一种变种。窗口大小是很灵活的。可以固定窗口大小滑动，也可以在滑动的过程中控制窗口大小。例如用for控制左边界，用while控制右边界。 
list set都可以使用==比较符号。滑动窗口时，可以设定一个预期哈希表，滑动过程中更新滑动哈希表与其进行比较。

## 切片赋值

需要注意的是，原地修改原数组的切片赋值方法为：

```python
nums = [1,2,3,4,5,6,7]
k = 3
n = len(nums)
nums[:] = nums[n-k:]+nums[:n-k]
```

而不是nums = nums[n-k:]+nums[:n-k] 。因为这样的操作本质是新建了一个局部变量nums，然后局部变量指向了这个新列表。
